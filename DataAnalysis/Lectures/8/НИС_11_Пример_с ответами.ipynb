# Задача 1
Выгрузить недавние новости с сайта https://nplus1.ru/ в датафрейм pandas и сохранить все в файл MS Excel.

```python
import requests
page = requests.get('https://nplus1.ru/')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
urls = [a['href'] for a in soup.find_all("a", href=True) if 'nplus1.ru/news' in a['href']]
urls
```

```python
new_0 = requests.get(urls[0])
soup0 = BeautifulSoup(new_0.text, 'html.parser')
soup0
```

```python
soup0.findAll('meta')
```

```python
soup0.find('meta', {'name': 'author'})
```

```python
soup0.find('meta', {'name': 'author'}).attrs
```

```python
soup0.find('meta', {'name': 'author'}).attrs['content']
```

```python
author = soup0.find('meta', {'name': 'author'})
title = soup0.find('meta', {'name': 'twitter:title'})
date_publ = soup0.find('meta', {'itemprop': 'datePublished'})

for tag in (author, title, date_publ):
    print(tag['content'])
```

```python
def GetUrls(url0):
    page0 = requests.get(url0)
    soup0 = BeautifulSoup(page0.text, 'html.parser')
    author = soup0.find('meta', {'name': 'author'}).attrs['content']
    title = soup0.find('meta', {'name': 'twitter:title'}).attrs['content']
    date = soup0.find('meta', {'itemprop': 'datePublished'}).attrs['content']
    return url0, date, author, title
```

```python
from time import sleep
news = []

for link in urls:
    res = GetUrls(link)
    news.append(res)
    sleep(1)
news[1]
```

```python
news
```

```python
import pandas as pd
df = pd.DataFrame(news)
df
```

```python
df.columns=['link', 'date', 'author', 'title']
df
```

```python
df.to_excel('mynews.xlsx')
```

# Задача 2
Скачаем все pdf, которые есть на странице про академический календарь НИУ ВШЭ. https://www.hse.ru/studyspravka/grafik

```python
a = requests.get('https://www.hse.ru/studyspravka/grafik')
soup = BeautifulSoup(a.text)
print(soup.prettify())
```

```python
links = soup.find_all('a')
pdf_links = []
for link in links:
    if link.get('href').endswith('.pdf'):
        pdf_links.append(link.get('href'))


pdf_links
```

```python
full_pdf_links = list(map(lambda x: 'https://www.hse.ru' + x, pdf_links))
full_pdf_links
```

```python
for link in full_pdf_links:
    r = requests.get(link)
    with open(link.split('/')[-1], 'wb') as f:
        f.write(r.content)
```
