```python
rows = page2.findAll('tr') #найдёт все объекты заданного типа
rows
```

```python
len(rows) #это список из нескольких элементов
```

```python
for i, row in enumerate(rows): #выводим пронумерованные строки
    print(i)
    print(row)
```

```python
for i, row in enumerate(rows): #содержимое первой ячейки каждой строки
    print(i)
    print(row.td.string)
```

```python
table = []#сохраним данные в виде списка списков
for row in rows:
    table.append([cell.string for cell in row('td')]) 
print(table)
```

# Поиск данных на HTML-странице¶

```python
url = "https://ru.wikipedia.org/wiki/%D0%9C%D0%B0%D1%81%D1%82%D0%B5%D1%80_%D0%B8_%D0%9C%D0%B0%D1%80%D0%B3%D0%B0%D1%80%D0%B8%D1%82%D0%B0"
```

```python
g = requests.get(url)
page = BeautifulSoup(g.text, "html.parser")
```

```python
for link in page('a'):
    print(link)
```

```python
for link in page.findAll("a", class_='external text'):
    print(link['href'])
```

### Проверьте себя

```python
html = """<html>
    <body>
        <p>
            Hello,
            <strong>
                World
            </strong>
        </p>
    </body>
</html>"""

get_strong(html).strip() == "World"
```
